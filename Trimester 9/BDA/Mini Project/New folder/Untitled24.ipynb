```python
random_orders__delivery_address_lat_lon__001 = []
random_orders__delivery_address_lat_lon__002 = []


key = '830323b5ca694362904814ff0a11b803'
geocoder = OpenCageGeocode(key)
```

```python
latlongeo = pd.DataFrame(data=data, columns=['random_orders__delivery_mobile_no', 'random_orders__delivery_address_lat_lon__002','random_orders__delivery_address_lat_lon__001'], index=None)
latlongeo
```

```python
mod_data = latlongeo.dropna(how='all', subset=['random_orders__delivery_address_lat_lon__002','random_orders__delivery_address_lat_lon__001'] )
mod_data
```

```python
address = 'India, Maharashtra, Pune'

location = geocoder.geocode(address)
random_orders__delivery_address_lat_lon__002 = location[0]['geometry']['lat']
random_orders__delivery_address_lat_lon__001 = location[0]['geometry']['lng']

print('The geograpical coordinate of Maharashtra, Pune are {}, {}.'.format(random_orders__delivery_address_lat_lon__002, random_orders__delivery_address_lat_lon__001))
```

```python
IN_map = folium.Map(location=[random_orders__delivery_address_lat_lon__002, random_orders__delivery_address_lat_lon__001], zoom_start=12)

# add markers to map
for lat, lng, linked_mobiles__ in zip(mod_data['random_orders__delivery_address_lat_lon__001'], mod_data['random_orders__delivery_address_lat_lon__002'], mod_data['random_orders__delivery_mobile_no']):
    label = '{}'.format(linked_mobiles__)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='red',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(IN_map)  
    
IN_map
```
